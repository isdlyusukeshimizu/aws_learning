```python
from byaldi import RAGMultiModalModel
# Optionally, you can specify an `index_root`, which is where it'll save the index. It defaults to ".byaldi/".
RAG = RAGMultiModalModel.from_pretrained("vidore/colpali-v1.2")
```

```python
RAG.index(
    input_path="/content/docs/", # The path to your documents
    index_name="paper", # The name you want to give to your index. It'll be saved at `index_root/index_name/`.
    store_collection_with_index=True, # Whether the index should store the base64 encoded documents. # Optionally, you can specify a list of document IDs. They must be integers and match the number of documents you're passing. Otherwise, doc_ids will be automatically created.
    overwrite=True, # Whether to overwrite an index if it already exists. If False, it'll return None and do nothing if `index_root/index_name` exists.
)
```

```python
results = RAG.search("Proposal Medthod?", k=3)
```

```python
results
```

```python
import base64
from io import BytesIO
from PIL import Image
import IPython.display as display

base_64s = []
for result in results:
    print(f"Doc ID: {result.doc_id}, Page: {result.page_num}, Score: {result.score}")
    print(f"Base64: {result.base64[:100]}...")
    base_64s.append(result.base64)

html_images = ""
for base64_data in base_64s:
    image_data = base64.b64decode(base64_data)
    image = Image.open(BytesIO(image_data))

    new_width = 300
    aspect_ratio = image.height / image.width
    new_height = int(new_width * aspect_ratio)
    resized_image = image.resize((new_width, new_height))

    buffered = BytesIO()
    resized_image.save(buffered, format="PNG")
    img_str = base64.b64encode(buffered.getvalue()).decode("utf-8")

    html_images += f'<img src="data:image/png;base64,{img_str}" style="display:inline-block; margin-right: 10px;" />'

display.display(display.HTML(f'<div style="white-space: nowrap;">{html_images}</div>'))
```

```python
import os
from pathlib import Path
from byaldi import RAGMultiModalModel

# Initialize RAGMultiModalModel
model = RAGMultiModalModel.from_pretrained("vidore/colqwen2-v0.1")
```

```python
import random
random_number = random.randint(1, 100)
```

```python
# Use ColQwen2 to index and store the presentation
index_name = "colpali_paper222"
model.index(input_path=Path("/content/Named Clinical Entity Recognition Benchmark/2410.05046.pdf"),
    index_name=index_name,
    store_collection_with_index=True, # Stores base64 images along with the vectors
    overwrite=True,
    doc_ids=[random_number]
)
```

```python
query = "本研究の提案手法は?"
results = model.search(query, k=5)
```

```python
import base64
from io import BytesIO
from PIL import Image
import IPython.display as display

base_64s = []
for result in results:
    print(f"Doc ID: {result.doc_id}, Page: {result.page_num}, Score: {result.score}")
    print(f"Base64: {result.base64[:100]}...")
    base_64s.append(result.base64)

html_images = ""
for base64_data in base_64s:
    image_data = base64.b64decode(base64_data)
    image = Image.open(BytesIO(image_data))

    new_width = 300
    aspect_ratio = image.height / image.width
    new_height = int(new_width * aspect_ratio)
    resized_image = image.resize((new_width, new_height))

    buffered = BytesIO()
    resized_image.save(buffered, format="PNG")
    img_str = base64.b64encode(buffered.getvalue()).decode("utf-8")

    html_images += f'<img src="data:image/png;base64,{img_str}" style="display:inline-block; margin-right: 10px;" />'

display.display(display.HTML(f'<div style="white-space: nowrap;">{html_images}</div>'))
```

```python
import base64
from io import BytesIO
from PIL import Image
import IPython.display as display
from openai import AzureOpenAI, OpenAI

query = "本研究の「研究背景」は何ですか？"
results = model.search(query, k=8)

base_64s = []
for result in results:
    print(f"Doc ID: {result.doc_id}, Page: {result.page_num}, Score: {result.score}")
    print(f"Base64: {result.base64[:100]}...")
    base_64s.append(result.base64)

openai_client = AzureOpenAI(
    azure_endpoint = os.getenv('ENDPOINT'), # 生成したリソースのエンドポイントです
    api_version = os.getenv('API_version'), # https://learn.microsoft.com/en-us/azure/ai-services/openai/reference を参照してバージョンを指定
    api_key = os.getenv('API_key') # 生成したリソースのキーです
)

response = openai_client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "system",
                "content": """
あなたは優秀な研究者です。
あなたのタスクは、論文の内容を分かりやすく日本語で解説することです。
""",
            },
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text":"""
# {役割}
以下の{背景}を理解し、論文の内容を分かりやすく、かつ具体的に説明してください。

# {背景}
時間が無い中でたくさんの論文を読む人のために、短時間で論文の内容をしっかりと理解できるような解説文章を作成する必要があります。

# {内容ルール}
- 網羅的に説明してください。
- 論文の「研究背景」と「提案手法」と「実験」と「結論」を出力内容に入れてください。
- 論文の重要な内容を含めてください。
- 700文字以上で出力してください。
- {ターゲット}向けに説明してください。
- {禁止事項}を守ってください。
- {用語統一ルール}を守ってください。

# {文体ルール}
- 「です・ます調」で出力してください。
- 句構造文法の枠組みを用いて、理解しやすい文章を作成してください。
- なるべく平易な言葉を用いて、自然な文章を出力することを心がけてください。
- 日本語で出力してください。
- 初心者でも理解できるように嚙み砕いて説明してください。

# {禁止事項}
- 箇条書きを多用すると、ペナルティを課します
- 見出し(#, ##, ###, ####)を出力に含めるとペナルティを課します
- 単に和訳した文章を出力すると、ペナルティを課します。
- {禁止ワード}に載っている単語を使用すると、ペナルティを課します。
- 推測によって事実と異なる文章を書くと、ペナルティを課します。
- ３文以上にわたって同じ語尾で続かせると、ペナルティを課します。（例：～ました。～ました。～ました。）
- Markdown記法を使用すると、ペナルティを課します。

# {ターゲット}
- 機械学習の入門を理解できる人
- 時間が無い人
- 時短で論文の内容をしっかりと理解したい人

# {用語統一ルール}
- トランスフォーマー→Transformer
- オーディオ→音声
- 大規模言語モデル→LLM
- LLMs→LLM

# {禁止ワード}
- これにより、
- 大言語モデル
""",
                    },
                ] + list(map(
                        lambda x: {
                            "type": "image_url",
                            "image_url": {
                                "url": f"data:image/jpg;base64,{x}",
                                "detail": "low",
                            },
                        },
                        base_64s,
                    ))
            },
        ],
    )
```
